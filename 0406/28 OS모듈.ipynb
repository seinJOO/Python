```python
## 특정 디렉터리 내에 있는 텍스트 자료 수집

import os

# 1. 텍스트 디렉터리 경로 지정
print(os.getcwd())          # 현재 기본 경로는 d:\sein\Python\0406 !
txt_data = 'data/txt_data/'

# 2. 텍스트 디렉터리 목록 반환
sub_dir = os.listdir(txt_data)
print(sub_dir)              # ['first', 'second']

# 3. 각 디렉터리의 텍스트 자료 수집 함수
def textPro(sub_dir) :      # 리스트의 first, second가 인자값으로 넘어감
    first_txt = []          # first 디렉터리 텍스트 저장
    second_txt = []         # second 디렉터리 텍스트 저장
    
    # 디렉터리 구성
    for sdir in sub_dir :
        dirname = txt_data + sdir       # 디렉터리 경로 구성 - 'data/txt_data/' + 디렉터리명
        file_list = os.listdir(dirname) # 디렉터리의 파일 리스트를 추려냄
        
        # 파일 구성
        for fname in file_list :
            file_path = dirname + '/' + fname   # 파일 경로 구성 = 'data/txt_data/' + 디렉터리명 + '/' + 파일명
            # 파일리스트 중 파일만 선택
            if os.path.isfile(file_path) :
                try :
                    file = open(file_path, 'r')
                    if sdir == 'first' :
                        first_txt.append(file.read())
                    else :
                        second_txt.append(file.read())
                except Exception as e :
                    print('예외 발생 !',e)
                finally :
                    file.close()
                    
    return first_txt, second_txt        # 텍스트 자료 반환

## 함수 호출
first_txt, second_txt = textPro(sub_dir)

## 수집한 자료 확인
print('first_txt 길이 :',len(first_txt))    # 파일이 10개인 걸 확인 가능 !
print('second_txt 길이 :',len(second_txt))

## 수집한 자료 결합
tot_txt = first_txt + second_txt
print('tot_txt 길이 :',len(tot_txt))     # 각 10개 씩 20개의 파일이 잘 결합됨!

#print('전체 텍스트\n',tot_txt)            # * 문자열로 합치고 싶다면 join 활용 !



## [문제 1] 위 데이터를 피클로 합쳐서 저장 후 읽어오기 - 저장 파일 : data/tot_texts.pck
import pickle

with open('data/tot_texts.pck','wb') as f :
    pickle.dump(tot_txt, f)
with open('data/tot_texts.pck','rb') as f :
    data = pickle.load(f)
# print(data)

# 쌤 답

# file save => pickle(binary)
sfile = open('data/tot_texts.pck','wb')
# pickle.dump()
pickle.dump(tot_txt, sfile)           # sfile에 있는 데이터(파일)에 pickle로 쓰기
# load file => pickle파일(data/tot_texts.pck)
rfile = open('data/tot_texts.pck','rb')
tot_texts_pck = pickle.load(rfile)      # rfile에 있는 데이터(파일)를 pickle로 읽어서 load

# 확인
print(type(tot_texts_pck))
    
```

```python
### 이미지파일만 선택하여 이동시키기
# data/images => data/images2
# [절차]
# 1. os.path.exists를 사용하여 디렉터리 유무 확인
# 2. glob(path/*.png) : png 파일 검색
# 3. png 이미지 복사는 이진 파일 읽기
# 4. png 이미지 파일 쓰기는 이진 파일 쓰기

import os
from glob import glob

imglist = glob('data/images/*.png')

for olddir in imglist :
    img = os.path.split(olddir)
    
    newpath = 'data/images2/'+img[1]
        
    if not os.path.exists('data/images2') :
        os.mkdir('data/images2')
        
    copy = open(olddir,'rb').read()
    open(newpath, 'wb').write(copy)
    
print(os.listdir('data/images2'))
        
```

```python
# 선생님 답
print(os.getcwd())
img_path = 'data/images/'
if (os.path.exists('data/images2')) :
    print('해당 디렉터리가 존재합니다')
else :
    os.mkdir('data/images2')
    print('data/images2 디렉터리를 생성합니다')
img_path2 = 'data/images2/'

# 디렉터리 생성/존재확인 후 image파일 작업 진행
if os.path.exists(img_path) :
    print('/data/images 디렉터리가 존재합니다')
    
    # 이미지 파일 작업    
    images = []
    for pic_path in glob(img_path+'*.png') :
        # 경로와 파일명 분리
        img_path = os.path.split(pic_path)      # 파일 이름을 변경해서 images2에 저장할 예정
        images.append(img_path[1])   
        # 이진파일로 읽기
        with open(pic_path,'rb') as rf :
            buf = rf.read()
        # 이진파일로 쓰기
        with open(img_path2+img_path[1],'wb') as wf :
            wf.write(buf)
else :
    print('해당 디렉터리가 존재하지 않습니다')        

print('copied png files :',images)
```
