```python
# 시간과 관련있는 모듈들
import datetime, time

# datetime 모듈은 시간에 대해서 가공된 정보를 처리
s = datetime.datetime.now()     # datetime 모듈 내의 datetime 클래스 내의 now 메서드
print(s)

t = time.localtime()            # time 모듈 내의 localtime 메서드
print(t)        # => 딕셔너리 형태로 데이터가 저징되어있음
# time.struct_time(tm_year=2022, tm_mon=3, tm_mday=29, tm_hour=10, tm_min=47, tm_sec=40, tm_wday=1, tm_yday=88, tm_isdst=0)
# tm_wday=1 => 요일은 월요일=0 부터 시작

print(t.tm_year)                    # 2022
print(time.localtime().tm_zone)     # 대한민국 표준시

start = time.time() # 1970.01.01.00시를 기점으로 초 계산을 함   (랜덤의 시드값이 됨)
print(start)        # 반환값 float      -> 1648519144.4199867
time.sleep(5)       # 5초 멈춘 뒤
end = time.time()   # 시간 다시 알아오기  
print(end)          #                  -> 1648519149.4269469
```

```python
# 시작점 만들기 (모듈)
# 모듈 안에 변수, 함수, 클래스 등을 포함할 수 있음 -> 이런 프로그램의 시작이 되는 시작점(main)을 만들 수 있음

# (형식)
# if __name__ == '__main__' :   # 프로그램 시작점(현재 동작하는 특수변수 __name__이 __main__일 때만 실행하도록 시작점 설정)
#       명령문(메인에서 실행할 코드)                       


# 연습 : 프로그램 시작점 만들기

# 1. 평균(mean)과 제곱근(sqrt) 모듈 import
from statistics import mean     # 평균
from math import sqrt           # 제곱근

# 2. 산출평균
def Avg(data) :
    avg = mean(data)
    return avg

# 3. 분산/표준편차      * 분산 : (개별 값 - 평균)의 제곱의 합 * 표준편차 : 분산의 제곱근
def var_sd(data) :  
    avg = Avg(data)     # 함수 호출
    diff = [ (d - avg) **2 for d in data]   # list 안에서 for문 실행시키기
    var = sum(diff) / (len(data) - 1)
    sd = sqrt(var)
    return var, sd

# 프로그램 시작점   => 만약 시작점 설정이 없다면 다른 파일에서 import 하자마자 실행됨
if __name__ == '__main__' :
    data = [1,3,5,7]
    print('평균 :',Avg(data))
    var, sd = var_sd(data)
    print('분산 :',var)
    print('표준편차 :',sd)
```
