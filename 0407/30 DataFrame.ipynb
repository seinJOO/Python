```python
# 공공데이터 파일 읽어보기
import pandas as pd

data = pd.read_csv(open('data/subway_complexity.csv'))

days = data['조사일자']
nums = data['호선']
name = data['역명']
per = data['9시00분']

days_5, num_5, name_5, per_5 = [],[],[],[]

for x in range(len(data)) :
    if nums[x] == '5호선' :
        days_5.append(days[x])
        name_5.append(name[x])
        per_5.append(per[x])

for x in range(len(days_5)) :
    if max(per_5) == per_5[x] :
        print('9시에 가장 혼잡한 5호선 역 :',days_5[x],name_5[x], '\n혼잡도 :',per_5[x])
```

```python
# 교통사고 파일을 열어서 DataFrame으로 생성
# 1. 생성된 DataFrame을 통해 총 사건수, 사망자수, 중상자수, 경상자수 조회
# 2. 서울 지역 사망자수를 조회하여 총 사망자수를 조회

data = pd.read_csv('data/교통사고 20201231.csv')
data = pd.DataFrame(data)

print('총 사건 수 :',sum(data['사고건수']))
print('총 사망자 수 :',sum(data['사망자수']))
print('총 중상자 수 :',sum(data['중상자수']))
print('총 경상자 수 :',sum(data['경상자수']))

dead_s = 0
for x in range(len(data)) :
    if data['시도'][x] == '서울' :
        dead_s += data['사망자수'][x]
print('서울 지역 총 사망자 수 :', dead_s)
```

```python
# 한국도로공사_휴게소 주차장 현황
# 1. 노선별 주차장의 합계를 딕셔너리로 데이터를 조회 생성
# 2. 각 본부별 주차장 수를 딕셔너리 데이터로 조회 생성
import pandas as pd
# data2 = pd.read_csv('data/한국도로공사_휴게소 주차장 현황_20201231.csv',encoding='euc-kr')
data2 = pd.read_csv(open('data/한국도로공사_휴게소 주차장 현황_20201231.csv'))
data2 = pd.DataFrame(data2)

zonePerLine = {}
for x in range(len(data2)) :
    zonePerLine[data2['노선'][x]] = zonePerLine.get(data2['노선'][x],0) + data2['합계'][x]
print(zonePerLine)

zonePerDept = {}
for x in range(len(data2)) :
    zonePerDept[data2['본부'][x]] = zonePerDept.get(data2['본부'][x],0) + data2['합계'][x]
print(zonePerDept)

# 문제 - comprehension을 이용하여 작성하기
lines = {
    data2['본부'][x] : sum(data2['합계'][y] for y in range(len(data2)) if data2['본부'][x] == data2['본부'][y]) for x in range(len(data2))
}
print(lines)

# ** 딕셔너리 내포
key = ['1','2','3','4']
val = [1,2,3,4]
test_dic = { key[i] : val[i] for i in range(4)}
# print(test_dic)     # {'1': 1, '2': 2, '3': 3, '4': 4}
```

```python
## Pandas : DataFrame의 구조 (= Table과 같은 구조)
#           Series들이 결합되어 있는 형태, Series 컬럼을 기반으로 데이터가 저장되어 있는 형태
#           Pandas에서 참조 시에는 Series의 이름을 사용하여 참조
## Series : List, Dictionary와 같은 구조
#           차이점 : 인덱싱 방식에 따라 리스트와 다른 인덱싱이 가능함 (리스트는 인덱스가 0부터 시작하지만, Series는 다르게 설정할 수 있음)

score = pd.read_csv('data/score.csv')
ls = [10,20,30,40,50]
print(ls)

s = pd.Series(ls,name='lst')
print(s[0])     # 10

# for문으로 list와 동일한 동작 확인
for i in s :
    print(i, end=' ')       # 10 20 30 40 50 => 리스트와 동일
print()
for r in range(len(s)) :
    print(s[r], end=' ')    # 10 20 30 40 50 => 리스트와 동일
print()

# 슬라이싱 확인
print(s[:2])                # 0    10 
                            # 1    20 => 슬라이싱도 가능

# 리스트와 다른 점 - 인덱싱 처리
s1 = pd.Series(ls, index=[1,2,3,4,5],name = 'index 수정')       # 인덱스 값은 튜플, 리스트, 딕셔너리로 만들 수 있다!
print(s1.info())

# 요구사항 - lotto1 이름의 시리즈를 만들어서 dic1 딕셔너리의 쌍을 가져와서 시리즈로 만들기
lotto1 = pd.Series([5,11,14,30,33,38,24], index=[1,2,3,4,5,6,'bonus'])
print(lotto1)

dic1 = {'1':5,'2':11,'3':14,'4':30,'5':33,'6':38,'bonus':24}
lotto2 = pd.Series(dic1)
print(lotto2)
```

```python
# 요구사항 - lotto1 이름의 시리즈를 만들어서 dic1 딕셔너리의 쌍을 가져와서 시리즈로 만들기
from random import randint

nums = []
while len(nums) < 7 :
    num = randint(1,45)
    if num not in nums :
        nums.append(num)
        
dic1 = {
    i+1 : nums[i] for i in range(len(nums)-1)
}
dic1.update({'bonus': nums[6]})
print(dic1)
```

```python
# 4개의 Series로 만들기
# 이름(name_s), 국어(kor_s), 영어(eng_s), 수학(math_s)

jumsu = {
    '홍길동' : [50,60,50],
    '이순신' : [60,60,60],
    '강감찬' : [100,90,90],
    '유관순' : [80,80,90],
    '강환석' : [95,95,95]
}

name = pd.Series(jumsu.keys(),name='name_s')
kor = pd.Series([x[0] for x in jumsu.values()], name='kor_s')
eng = pd.Series([x[1] for x in jumsu.values()], name='eng_s')
math = pd.Series([x[2] for x in jumsu.values()], name='math_s')

score = pd.DataFrame({'이름' : name, '국어' : kor, '영어' : eng, '수학' : math})
print(score)

total = kor + eng + math
print(total)
avg = round(total/3,2)
print(avg)

# 참고 : 인덱스 값이 같아야 잘 합쳐진다 !!!!
score['합계'] = total
score['평균'] = avg
score
```
