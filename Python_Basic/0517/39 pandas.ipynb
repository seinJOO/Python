```python
## 판다스 연습

# sample_1.xlsx 데이터 불러오기
import pandas as pd

sample_1 = pd.read_excel('data/sample_1.xlsx',
                         header=1,
                         skipfooter=2,
                         usecols="A:C")
sample_1.head(3)    # 상위에서부터 (헤더를 제외한) 출력 데이터 개수 지정
sample_1.tail(3)    # 하위에서부터 (헤더를 제외한) 출력 데이터 개수 지정
```

```python
## sample_1의 데이터 정보 조회 - info()
sample_1.info()
```

```python
## sample_1의 데이터 기초통계량 확인 : describe()
sample_1.describe()
```

```python
# 연습문제 - 각 컬럼 별 시리즈를 확인하기
# 입국객수 / 국적코드, 입국객수 / 
print(sample_1['입국객수'], type(sample_1['입국객수']))
print(sample_1[['국적코드','입국객수']], type(sample_1[['국적코드','입국객수']]))
print(sample_1[['국적코드','성별']])

# 컬럼 생성하기
# sample_1['기준연월'] = pd.Series(['2019-11' for x in range(len(sample_1))],name='기준연월')
sample_1['기준연월'] = '2019-11'
sample_1['test'] = ['2019-11','2019-12','2020-01','2020-02','2020-03','2020-04']
sample_1.drop(columns="test")
```

```python
## 필터링하기

# 1. 성별이 남성인 데이터 출력
conditions = (sample_1['성별']=='남성')
sample_1[conditions]

# 2. 입국객수가 150000을 넘기는 경우만 출력
conditions = (sample_1['입국객수']>150000)
sample_1[conditions]
# 3. 성별이 남성이면서 150000을 넘기는 경우만 출력
conditions = ((sample_1['성별']=='남성') & (sample_1['입국객수']>150000))
sample_1[conditions]
```

```python
# 한 컬럼(시리즈)에서 여러 조건으로 필터링
# 국적코드가 A01, A18인 경우만 출력

conditions = ((sample_1['국적코드'] == 'A01')|(sample_1['국적코드'] == 'A18'))
sample_1[conditions]

conditions = (sample_1['국적코드'].isin(['A01','A18']))
sample_1[conditions==False]     # False인 것만 출력
```

```python
sample_1 = sample_1.drop(columns='test')
```

```python
# 데이터 합치기

# sample_codemaster.xlsx 데이터프레임으로 불러오기
code_master = pd.read_excel('data/sample_codemaster.xlsx')
code_master
```

```python
# 옆으로 통합하기
sample_1_code = pd.merge(left=sample_1, right=code_master)
sample_1_code   # code_master에 존재하지 않는 국적 코드 데이터는 함께 저장되지 않는다 !

sample_1_code = pd.merge(left=sample_1, right=code_master,how='left')
sample_1_code   # how='left'일 때 LEFT OUTER JOIN과 같은 효과 !

sample_1_code = pd.merge(left=sample_1, right=code_master,how='inner',left_on='국적코드',right_on='국적코드')
sample_1_code   # merge의 기준을 따로 지정할 수 있음!! (단, left_on과 right_on을 함께 써야함)
```

```python
sample_2 = pd.read_excel('data/sample_2.xlsx', header=1, skipfooter=2, usecols='A:C')
sample_2['기준연월'] = '2019-12'

sample_2_code = pd.merge(sample_2, code_master, how='left', left_on='국적코드', right_on='국적코드')
sample_2_code
```

```python
# 아래로 통합하기
sample = sample_1_code.append(sample_2_code, ignore_index=True)
sample
```

```python
### pandas.concat() 사용하여 아래로 병합하기
sample2 = pd.concat([sample_1_code,sample_2_code],ignore_index=True)
sample2
```

```python
# 데이터를 엑셀파일로 저장하기
sample.to_excel('data/sample.xlsx')
```

```python
# 인덱스값을 제거하여 저장하기
sample2.to_excel('data/sample2.xlsx',index=False)
```

```python
### 데이터 집계하기 (pivot table)
# pivot_table() 함수 사용 - 기준연월 당 국적 별 입국객수의 평균
sample_pivot = sample.pivot_table(values='입국객수', index='국적명', columns='기준연월', aggfunc='mean')
sample_pivot
```

```python
# 전체 기간의 국적 별 입국객수의 평균
sample_pivot = sample.pivot_table(values='입국객수', index='국적명', aggfunc='mean')
sample_pivot
```

```python
# 전체 기간의 국적 별 입국객수의 합계(max, sum, min, mean, ...)
sample_pivot = sample.pivot_table(values='입국객수', index='국적명', columns='기준연월', aggfunc='sum')
sample_pivot
```
