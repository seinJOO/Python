```python
## 자료 시각화 (matplotlib 패키지 설치)
# 가장 전통적인 시각화 모델 - pyplot 모듈
# pip install matplotlib
```

```python
top_5 = [('지구촌', 10), ('인터뷰', 5), ('돋보기', 7), ('미러', 3), ('갈등', 2)]

# 단어와 출현 빈도수 만들기
words = []       # 단어
counts = []     # 빈도수

for word, cnt in top_5 :        # top_5가 튜플 형태이기 때문에 unpacking 후 list에 packing
    words.append(word)
    counts.append(cnt)

print(words)
print(counts)

import matplotlib.pyplot as plt     # 모듈 사용 시 별칭 사용

# 차트에 한글 적용을 위해, 한글을 지원하는 폰트 가져오기
from matplotlib import font_manager, rc
    # 적용시킬 폰트 이름 가져오기
font_name = font_manager.FontProperties(fname='C:\Windows\Fonts\malgun.ttf').get_name()
    # 폰트 이름으로 적용시키기
rc('font',family=font_name)

# 선 그래프 (plot)
plt.plot(words, counts)
plt.show()

# 막대 그래프 (bar)
plt.bar(words, counts)
plt.show()

# 원형 그래프 (pie)
plt.pie(counts, labels=words, autopct='%1.1f%%',textprops={'color' : 'w','fontsize':15})
plt.show()
```

```python
# [실습] data.pickle1로 직접 데이터 시각화하기

import pickle

file = open('data/data.pickle1',mode='rb')
news = pickle.load(file)
from data.textclean import cleaning
lines = [cleaning(row) for row in news]

word_list = {}

for line in lines :
    for word in line.split() :
        if len(word) > 1 :
            word_list[word] = word_list.get(word,0) + 1

del word_list['이번엔']     # 불필요한데 빈도수 높은 값 직접 제거

from collections import Counter
counter = Counter(word_list)
top5 = counter.most_common(5)

words = []
counts = []
for x, y in top5 :
    words.append(x)
    counts.append(y)

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='C:\Windows\Fonts\malgun.ttf').get_name()
rc('font',family=font_name)

import matplotlib.pyplot as plt
plt.plot(words, counts)
plt.show()
plt.bar(words, counts)
plt.show()
plt.pie(counts,labels=words,autopct='%1.1f%%',textprops={'color' : 'w','fontsize' : 15})
plt.show()
```
