```python
# 데이터 셋 분할
# 모델학습에 사용할 훈련 (train)용 데이터와 모델 성능을 평가하는데 사용할 검증용(test) 데이터로 분할함
# 보통 훈련 80% 테스트 20% 로 분할
# 전체 150개의 데이터를 테스트용 30개 훈련용 120개로 분할해보자 !!!

from sklearn.model_selection import train_test_split # 데이터셋 분할기능 가져오기
# 관용적으로 X는 대문자, y는 소문자로 사용함
X_data = df.loc[ : , 'sepal_length' : 'petal_width']   #열 데이터 순서대로 넣기 / 가장 결과값과 상관관계가 높은 petal_width는 필수로 넣어줘야함
y_data = df.loc[ : , 'Target']       #location 함수 .loc[ 행 , 열 ] => from : to 인데,  : 만 입력하면 행/열 전체

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.2, shuffle=True, random_state=20)
 #test_size : 검증용 데이터 20프로는 남기고 잘라주기 / shuffle : 데이터를 섞은 후 잘라주기 / random_state 랜덤값을 지정해주면 추출할 때마다 동일한 값을 추출함
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#KNN 알고리즘 : K-Nearest-Neighbors = K 근접이웃
# K값에 따라 모델의 예측력이 달라지므로 적절한 K값을 설정해야 한다

# 훈련용으로 그래프를 그렸을 때 검증용 데이터의 위치가 떨어지면 -> 그 데이터의 위치 근처에 있는 데이터들의 클래스를 따져본다
# 만약 근처에 결과값 A를 가진 데이터 하나와 결과값 B를 가진 데이터 두 개가 있을 경우 검증용 데이터는 결과값이 B를 가진다고 보면 된다
# K값으로 범위를 정하게 되는데, 각각 범위를 다르게 지정하여 훈련용 데이터의 결과와 검증용 데이터의 결과를 비교하여 K값을 확정하는 방식으로 한다

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)        # 알고리즘객체 knn을 생성하면서 생성자에 k값을 던져줌

knn.fit(X_train, y_train) # 훈련용데이터(문제지X, 답안지y)들로 공부 끝!

y_knn_pred = knn.predict(X_test)      # 검증용 데이터를 던져줘서 풀게 함

from sklearn.metrics import accuracy_score #정확도 기능 뽑아주기
knn_acc = accuracy_score(y_test, y_knn_pred) #정확도가 얼마인지 확인 -> 테스트 데이터를 통한 정답지와 검증용 데이터를 풀고난 정답지를 비교함
knn_acc

 # n = 2 acc = 0.9 / n = 7 acc =1.0 / n = 30 acc = 0.93
# n값과 정확도가 꼭 비례하지는 않는다!
```

```python
# SVM support Vector Machine
# 각 군집간의 거리(margin)를 최대한 멀리 유지하는 경계면을 찾기

from sklearn.svm import SVC   #
svc = SVC(kernel='rbf')           # 항상 같은 구문!
svc.fit(X_train, y_train)             # 훈련용 데이터로 군집간 거리 측정을 공부시킴

y_svc_pred = svc.predict(X_test)   # 검증용 데이터(문제)를 제공하여 답안지 만들게 하기
svc_acc = accuracy_score(y_test, y_svc_pred)  # 실제 검증용 데이터 답안지와 svc로 만든 답안지의 정확도 확인
svc_acc
```

```python
# Logistic Regression

# 각 품종에 속하는 확률을 계산하여 1에 가까우면 그 클래스로 분류, 0에 가까우면 아니라고 분류

from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression()      #Logistic Regression은 인수나 파라미터가 없음
lrc.fit(X_train, y_train)           # 훈련용 데이터로 공부시키기
y_lrc_pred = lrc.predict(X_test)    #검증용 데이터로 답안지 만들게 하기
lrc_acc = accuracy_score(y_test,y_lrc_pred) # 실제 검증용 데이터 답안지와 컴퓨터가 만든 검증용 데이터 답안지의 정확도 확인
lrc_acc
```

```python
# Decision Tree 의사결정나무

#트리의 각 분기점에는 데이터셋의 피처를 하나씩 위치시켜서, 해당 피처에 대한 임의의 조건식을 가지고 계속 2개의 줄기로 가지를 나누면서 데이터를 구분함
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth = 3, random_state = 20)   #깊이를 3으로 지정(의사결정을 최대 세 번 함) = 모델의 복잡도
dtc.fit(X_train, y_train)
y_dtc_pred = dtc.predict(X_test)
dtc_acc = accuracy_score(y_test,y_dtc_pred)
dtc_acc
```

```python
# 앙상블 - 보팅, 배깅, 부스팅

# 보팅 Voting : 하나의 모델 예측에 의존하는 것
# 여러 모델의 예측을 종합하면 예측력이 더 좋아짐 (각 모델들이 예측한 답으로 다수결에 따라 정해버림~!)

from sklearn.ensemble import VotingClassifier

hvc = VotingClassifier(estimators = [ ( 'KNN', knn ) , ( 'SVM', svc ) , ( 'DT', dtc ) ] , voting='hard' )
# voting 방식에는 hard : 표를 던지는 식 soft : 평균을 구하는 식 -> hard가 더 정확하다고 봄

hvc.fit(X_train, y_train)

y_hvc_pred = hvc.predict(X_test)
hvc_acc = accuracy_score(y_test, y_hvc_pred)
hvc_acc
```

```python
# 앙상블 - 배깅
# 한 알고리즘을 여러 번 돌리는 것

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 50, max_depth = 5 , random_state = 20)     # 위에 설정된 estimators들을 50번 돌리기
rfc.fit(X_train, y_train)
y_rfc_pred = rfc.predict(X_test)
rfc_acc = accuracy_score(y_test, y_rfc_pred)
rfc_acc
```

```python
# 앙상블 - 부스팅
# XGBoost 모델은 캐글, 데이콘 등 경진대회에 가장 많이 사용되는 알고리즘
# 모델 학습 속도가 빠르고 예측력이 상당히 좋은 편으로 알려져 있음

# 여러 개의 모델을 순차적으로 학습한다 - 오차를 줄일 수 있는 방향으로 모델을 계속 업데이트함
# 여러 모델을 동시에 학습하지 않고, 순차적으로 학습하는 형태 <-> 보팅, 배깅

# 배깅 = 모델들의 예측값을 종합 -> 결론
# 부스팅 = 모델1 예측 -> 모델1 예측값 + 모델 2 예측 ->  모델 1,2 예측값 + 모델 3 예측 -> 최종 예측

from xgboost import XGBClassifier
xgbc = XGBClassifier(n_estimators = 50, max_depth = 3 , random_state = 20)
xgbc.fit(X_train, y_train)
y_xgbc_pred = xgbc.predict(X_test)
xgbc_acc = accuracy_score(y_test,y_xgbc_pred)
xgbc_acc
```

```python
# 과대적합 overfitting : 훈련데이터와 관련된 예측은 원활하게 이루어지지만 조금 다른 데이터는 잘 인식하지 못하는 현상
# 과소적합 underfitting : 훈련데이터가 충분하지 못해서 예측을 잘 못하는 현상
# => 과대적합은 조금 다른 데이터의 양을 늘려주고, 과소적합은 훈련데이터의 양을 늘려주는 방식으로 해결한다 !
# hold out : 데이터 양 뻥튀기! 훈련데이터에서 또 잘라내어서, 나머지 훈련데이터로 학습을 시키고 잘라낸 훈련데이터로 검증을 한다
```

```python
# Hold-out 교차검증

# X_train y_train : 훈련용 문제와 훈련용 정답 (80%) -> 얘를 훈련용/검증용으로 다시 잘라보자~! => X_tr , X-val , y_tr, y-val
# X_test y_test : 검증용 문제와 검증용 정답 (20%)

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train,
                                              test_size = 0.3,
                                              shuffle = True,
                                              random_state = 20
                                              )      # 30프로를 검증용, 70프로를 훈련용으로 다시 나누기

rfc = RandomForestClassifier(max_depth = 3, random_state=20)
rfc.fit(X_tr, y_tr)
y_tr_pred = rfc.predict(X_tr)
y_val_pred = rfc.predict(X_val)
tr_acc = accuracy_score(y_tr, y_tr_pred) #훈련용 데이터의 학습을 통한 예측치와, 훈련용 데이터 내의 훈련용 데이터 학습을 통한 예측치 비교?
val_acc = accuracy_score(y_val, y_val_pred) #검증용 데이터의 예측치와, 훈련용 데이터 내의 검증용 데이터를 통한 예측치 비교?

print(tr_acc, val_acc)
# tr_acc = 1.0       val_acc = 0.9166666666666666
#검증용 정확도가 0.91로 낮은 수준 - 훈련용 데이터에 과대적합하여 새로운 데이터에 대한 예측력이 떨어지고 있음
```

```python
y_test_pred = rfc.predict(X_test)
test_acc = accuracy_score(y_test, y_test_pred)
test_acc
```

```python
# k-fold

# 홀드아웃은 학습 데이터를 다시 훈련용과 검증용으로 한 번만 분할하여 성능을 검증
# k-fold는 홀드아웃을 여러번 반복하는 방법
# 데이터를 서로 겹치지 않게 분할하여 총 k번 검증하므로 모델의 일반화 성능을 평가하는 데 뛰어남

from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, shuffle = True, random_state = 20) #5개로 분할함

num_fold = 1
# 분할한 데이터를 확인해보즈아
for tr_idx, val_idx in kfold.split(X_train) :
  print("%s Fold ----------------------------" % num_fold)
  print("훈련용 : ", len(tr_idx), tr_idx[:23])    #인덱스 처음부터 23까지만 뽑기
  print("검증용 : ", len(val_idx), val_idx[:23])
  num_fold += 1

  # 훈련용 데이터는 중복되나 검증용 데이터는 중복되지 않음
  # 분할하여 학습하였기 때문에 각 예측치가 모두 다르고, 가장 정확도가 높은 모델을 기본 모델로 설정할 수 있게 됨
```

```python
# 각 fold로 모델을 만들어 예측하고 정확도를 구해보자

from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, shuffle = True, random_state = 20)

val_scores = []
num_fold = 1
# X_train과 y_train은 처음에 80:20으로 분할했던 데이터들
# train데이터를 70:30으로 분할한 데이터들을 - 인덱스 값을 행(기준)으로 가져오고, 나머지(데이터)를 모두 열로 가져오기
for tr_idx, val_idx in kfold.split(X_train, y_train) :
  X_tr, X_val = X_train.iloc[tr_idx, :] , X_train.iloc[val_idx,:]    
  y_tr, y_val = y_train.iloc[tr_idx] , y_train.iloc[val_idx]
  #학습
  rfc = RandomForestClassifier(max_depth=5,random_state=20)
  rfc.fit(X_tr, y_tr)
  #검증
  y_val_pred = rfc.predict(X_val)
  val_acc = accuracy_score(y_val, y_val_pred)   #5번 분할한 데이터들의 모든 예측값의 정확도 비교
  print("%d Fold Accuracy : %.4f" %(num_fold, val_acc))
  val_scores.append(val_acc)
  num_fold += 1

```
