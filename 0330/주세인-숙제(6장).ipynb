```python
# [문제 1] Rectangle 클래스를 작성하시오
# 1. 멤버변수 : 가로, 세로
# 2. 생성자 : 가로, 세로 멤버 변수 초기화
# 3. 메서드 :  사각형의 넓이를 구하는 함수
# 4. 메서드 : 사각형의 둘레를 구하는 함수

print('사각형의 넓이와 둘레를 계산합니다')
w = int(input('사각형의 가로 입력 : '))
h = int(input('사각형의 세로 입력 : '))

class Rectangle :
    width = height = 0
    def __init__(self,width,height) :
        self.width = width
        self.height = height
    def area_calc(self) :
        return self.width * self.height
    def circum_cal(self) :
        return (self.width + self.height) * 2

rec = Rectangle(w, h)
print('='*30)
print('사각형의 넓이 :',rec.area_calc())
print('사각형의 둘레 :',rec.circum_cal())
print('='*30)
```

```python
# [문제 2] 동적 멤버 변수를 생성하여 다음과 같은 산포도를 구하는 클래스를 정의하시오 (분산, 표준편차)

from statistics import mean
from math import sqrt

x = [5,9,1,7,4,6]

# 산포도 클래스
class Scattering :
    nums = []
    # 생성자
    def __init__(self,data) :
        self.nums = data
    # 메서드 : 분산
    def var_func(self) :
        var = sum([ (d - mean(self.nums)) **2 for d in self.nums]) / (len(self.nums)-1)
        return var
    # 메서드 : 표준편차
    def std_func(self) :
        std = sqrt(sum([ (d - mean(self.nums)) **2 for d in self.nums]) / (len(self.nums)-1))
        return std

scat = Scattering(x)
print('분산 : {} \n표준편차 : {}'.format(scat.var_func(),scat.std_func()))
        
```

```python
# [문제 3] Person 클래스 작성
# 1. 멤버변수 : 이름, 성별, 나이
# 2. 생성자 : 이름, 성별, 나이 초기화
# 3. 메서드 : display(이름,성별,나이 출력)

class Person :
    name = None
    gender = None
    age = 0
    def __init__(self,age,name,gender) :
        self.name = name
        if gender == 'female' : self.gender = '여자'
        else : self.gender = '남자'
        self.age = age
    def display(self) :
        print('이름 : {}, 성별 : {} \n나이 : {}'.format(self.name,self.gender,self.age))

name = input('이름 입력 : ')
age = input('나이 입력 : ')
gender = input('성별(male/female) 입력 : ')
print('='*30)
p = Person(age, name, gender)
p.display()
print('='*30)
```

```python
# [문제 4] Employee 클래스를 상속받아서 Permanent와 Temporary 클래스를 구현
# 1. 키보드로 정규직과 임시직을 구분
# 2. 정규직인 경우에는 기본급과 상여금을 입력 받아서 급여 계산
# 3. 임시직인 경우에는 작업시간과 시급을 입력 받아서 급여 계산

class Employee :
    name = None
    pay = 0
    
    def __init__(self,name) :
        self.name = name
    
class Permanent(Employee) :
    base = 0 ; bonus = 0 ; contract = None
    
    def __init__(self, name, base, bonus) :
        self.name = name
        self.base = base
        self.bonus = bonus
        self.pay = base + bonus
        self.contract = '정규직'
    
    def display(self) :
        print('고용형태 : {} \n이름 : {} \n급여 : {:,}'.format(self.contract,self.name,self.pay))

class Temporary(Employee) :
    time = 0 ; wage = 0 ; contract = None
    
    def __init__(self, name, time, wage) :
        self.name = name
        self.time = time
        self.wage = wage
        self.pay = time * wage
        self.contract = '임시직'
    
    def display(self) :
        print('고용형태 : {} \n이름 : {} \n급여 : {:,}'.format(self.contract,self.name,self.pay))
        

empType = input('고용형태 선택(정규직<P>, 임시직<T>) : ')
if empType == 'P' or empType == 'p' :
    name = input('이름 : ')
    base = int(input('기본급 : '))
    bonus = int(input('상여급 : '))
    p = Permanent(name, base, bonus)
    p.display()

elif empType == 'T' or empType == 't' :
    name = input('이름 : ')
    time = int(input('작업시간 : '))
    wage = int(input('시급 : '))
    t = Temporary(name, time, wage)
    t.display()

else : 
    print('='*30)
    print('입력 오류')
```
